```python
# ============================================================
# Sport England Active Lives Analysis
# Active Adults Survey 2015-2024
# ============================================================

from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt

# -- UPLOAD FILE --
uploaded = files.upload()

# -- LOAD DATA --
df = pd.read_excel('REPORT Nov 23-24_Table 1-5_Levels of activity v2 - corrected.xlsx',
                   sheet_name='Table 1b Levels Trends',
                   header=None)

# -- EXTRACT DATA --
years = [
    '2015-16', '2016-17', '2017-18', '2018-19', '2019-20',
    '2020-21', '2021-22', '2022-23', '2023-24'
]

active_rate_cols = [3, 11, 19, 27, 35, 43, 51, 59, 67]

rows = {
    'All Adults': 7,
    'Male': 10,
    'Female': 11,
    '16-24': 15,
    '25-34': 16,
    '35-44': 17,
    '45-54': 18,
    '55-64': 19,
}

data = {}
for label, row in rows.items():
    data[label] = [pd.to_numeric(df.iloc[row, col], errors='coerce') for col in active_rate_cols]

results = pd.DataFrame(data, index=years)
age_groups = ['16-24', '25-34', '35-44', '45-54', '55-64']

# -- CHART 1: Overall trend --
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(years, results['All Adults'], marker='o', linewidth=2.5, color='#2ecc71')
ax.set_title('Adult Physical Activity Levels in England (2015–2024)', fontsize=14, fontweight='bold')
ax.set_ylabel('Active Rate (150+ mins/week)')
ax.set_xlabel('Year')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))
ax.axvspan('2020-21', '2021-22', alpha=0.15, color='red', label='COVID period')
ax.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('chart1_overall_trend.png', dpi=150)
plt.show()

# -- CHART 2: Gender gap --
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(years, results['Male'], marker='o', label='Male', color='#3498db', linewidth=2.5)
ax.plot(years, results['Female'], marker='o', label='Female', color='#e74c3c', linewidth=2.5)
ax.set_title('Activity Levels by Gender (2015–2024)', fontsize=14, fontweight='bold')
ax.set_ylabel('Active Rate (150+ mins/week)')
ax.set_xlabel('Year')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))
ax.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('chart2_gender_gap.png', dpi=150)
plt.show()

# -- CHART 3: Age groups latest year --
latest_rates = [results.loc['2023-24', age] for age in age_groups]
fig, ax = plt.subplots(figsize=(9, 5))
bars = ax.bar(age_groups, latest_rates, color='#9b59b6', edgecolor='white')
ax.set_title('Activity Rates by Age Group — England 2023–24', fontsize=14, fontweight='bold')
ax.set_ylabel('Active Rate (150+ mins/week)')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))
for bar, rate in zip(bars, latest_rates):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005,
            f'{rate:.1%}', ha='center', fontsize=11)
plt.tight_layout()
plt.savefig('chart3_age_groups.png', dpi=150)
plt.show()

# -- CHART 4: Age group trends --
fig, ax = plt.subplots(figsize=(11, 6))
colors = ['#e74c3c', '#e67e22', '#f1c40f', '#2ecc71', '#3498db']
for age, color in zip(age_groups, colors):
    ax.plot(years, results[age], marker='o', label=age, color=color, linewidth=2)
ax.set_title('Activity Trends by Age Group (2015–2024)', fontsize=14, fontweight='bold')
ax.set_ylabel('Active Rate (150+ mins/week)')
ax.set_xlabel('Year')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))
ax.legend(title='Age Group')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('chart4_age_trends.png', dpi=150)
plt.show()

# -- KEY FINDINGS --
print("""
KEY FINDINGS
============
1. Adult activity levels reached a record high of 63.7% in 2023-24
2. COVID caused a notable dip in 2020-21 before a strong recovery
3. Men are consistently ~5 percentage points more active than women
4. Activity declines steadily with age — 16-24s most active at 73%
""")
```
